```python
X = data.drop(labels = ['default payment next month'],axis=1)
```

```python
y = data['default payment next month']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y, test_size= 0.25 , random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train_scaled,y_train)
```

```python
y_pred = clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
score = accuracy_score(y_test,y_pred)
print(score)
```

## Cross validation

```python
param_grid = {"var_smoothing": [1e-9,0.1, 0.001, 0.5,0.05,0.01,1e-8,1e-7,1e-6,1e-10,1e-11]}
```

```python
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(clf , param_grid= param_grid , cv = 5, verbose= 3)
```

```python
gs.fit(X_train_scaled,y_train)
```

```python
gs.cv_results_
```

```python
gs.best_params_
```

```python
clf2 = GaussianNB(**gs.best_params_)
```

```python
clf2.fit(X_train_scaled,y_train)
y_pred2 = clf2.predict(X_test_scaled)
accuracy_score(y_test,y_pred2)*100
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
models = {"Logistic Regression":LogisticRegression(), "SVC": SVC(), "SGD": SGDClassifier(),"KNN":KNeighborsClassifier(), "Decision Tree":DecisionTreeClassifier(), "GBC": GradientBoostingClassifier(), "GNB":GaussianNB(),"Random Forest":RandomForestClassifier(),"XGB":XGBClassifier()}
```

```python
def training(models:dict, X_train_scaled , y_train , X_test_scaled , y_test):
    result = {}
    for i in models:
        model = models[i]
        model.fit(X_train_scaled,y_train)
        y_pred = model.predict(X_test_scaled)
        accuracy = accuracy_score(y_pred, y_test)
        result[model] = accuracy
    return result
        
training(models, X_train_scaled , y_train , X_test_scaled , y_test)
```
